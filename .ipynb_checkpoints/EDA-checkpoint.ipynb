# Exploring the Timeseries Dataset

## Problem Statement:  Which five zip codes are the best?

Our problem statement is very opent this time around. From our dataset of n counties, we're supposed to find the five 'best ones' for an investment firm. So, to get there, we're going to have to make a few assumptions:

* **Defining 'Best'.** The 'best' zip codes should probably be fairly low risk and moderate to high reward. We can calculate ROI based on a final projected value, and evaluate potential risk by evaluating the worst case final projected value.

* **Time horizon.** For a moderate confidence, we will look at just 5 years into the future.

So, since we've gotten a grip on how we want to approach our problem, let's take a look at the data we have.

### Making some assumptions about the data:

It looks like our data really does build itself out over time.

**Why?** If you review how zip codes work, which is what our database appears to be based off of, we find that zip codes are, in part, an addressing system local to a post office. So, in short, if a post office closes or opens, a zip code may be created, removed or consolidated, and boundaries are simply redrawn to accomodate. This is because zip codes are discrete objects, and not naturally occurring phenomena.

This implies that the NA values that we see are not true missing data - instead, it represents changes to the logistics and mail processing of USPS. Zip codes that exist in 2018, at the time this was pulled, may not have existed in the past as new post offices are established, consolidated, or closed. That's why we see fewer null values closer to the present, and more null values farther into the past.

This is important because, in our pre-processing step, we usully have to ask questions about how to handle NA values. In this case, extrapolating NA values for these zip codes would have a unique problem in that **the data we're extrapolating is already accurately captured elsewhere in the dataset.** As such, we won't be filling in any missing data - instead, we're going to drop NA values for each timeseries to **1) avoid any modeling risks w/ NA values, but 2) to also avoid creating false data that misrepresents the value of an existing zip code.**

### Let's start looking for other trends

Let's just drop whatever information we don't need from our dataframe.

```python
drop = list(df.columns[[0,2,3,4,5,6]])
df = df.drop(drop, axis = 1)
df.head()
```

There's no way we can individually visually inspect all 14.7K counties - so, we'll have to try and see if visually inspecting the overall trend can give us some insight. Let's start by looking at average values across time and evaluate this using an autocorrelation function.

```python
averages = []
for i in df.columns[1:]:
    averages.append(np.mean(df[i].dropna().values))

fig = plt.figure(figsize=(15,10))
plt.title('Average House Values in Dollars vs Time', fontsize = 20)
plt.plot(averages)
plt.ylabel('Average House Value in Dollars')
plt.xlabel('Date')
plt.xticks(ticks = [0,24,48,72,96,120,144,168,192,216,240,264],
          labels = df.columns[1:]);
```

### Let's examine autocorrelation for our data

First, we're going to convert our average series info into a dataframe for ease of plotting.

```python
keys = df.columns[1:]
values = averages
boop = dict(zip(keys,values))

df_average = pd.DataFrame.from_dict(boop, orient = 'index')
df_average.columns = ['Value']
df_average.index = pd.to_datetime(df_average.index)
df_average.head()
```

```python
plt.figure(figsize = (15,5))
pd.plotting.autocorrelation_plot(df_average);
```

Since `pandas` extrapolates all the way to 300, we can see if the partial autocorrelation function can help elucidate what we're looking at. Looks like there's a lot of changes to significance as time goes on, so let's try and visualize this some more with more control.

```python
rcParams['figure.figsize'] = 15,5
plot_pacf(df_average, lags = 300);
```

On review, it looks like there are strong correlations much farther back in the past than what initially makes sense.

### What does this mean?

Really valuable here is the initial plot of the total average trend. The housing market crash of 08/09 is visible here, and it makes sense that as houses that *lose* and *recover* their value, there's going to be much stronger correlation between their 'present' value and a distant past value, reflecting their overall decline and recovery. This would persist only until the market has fully recovered and properties are no longer trying to their past maximum.

**This does NOT mean** that we should model as far back as 200 periods to inform our prediction.

**However** what if we were looking at data that we didn't know the history on? We made the mistake of first looking at autocorrelation of non-stationary data. So, in cases where we aren't sure, we can always transform the data to a stationary set and evaluate it.

## Applying stationarity - bringing in a helper function

In the other notebook .[%Functions_and_Model_Testing], I have created a function that will help us simultaneously try out multiple transforms on our data, and then return the most stationary one.

```python
def best_transform(df):
    #This function will evaluate a given pandas timeseries Dataframe (n x 1) across multiple transforms for stationarity,
    #and then return the most stationary TS, along with the descriptors p for Dickey-Fuller test,
    #a string description of the type of the best transformation, and
    #a stat describing the nth order of the transform where applicable.
    
    ###
    #Create Pandas core series from (n x 1) dataframe
    TS = df[df.columns[0]]
    
    ###
    #Evaluate TS without transformation
    #Store Dickey-Fuller(DF) p value
    dftest = adfuller(TS.values)
    p_as_is = dftest[1]
    ###
    
    ###
    #Stationarity of ln transform
    #Store Dickey-Fuller(DF) p value
    dftest = adfuller(np.log(TS.values))
    p_log = dftest[1]
    ###
    
    ###
    #Stationarity of curve w/ rolling mean subtracted
    #Checks rolling mean across iterations of 2 - 6 periods
    #Selects best DF p value
    roll_result = {}
    roll_param = {}
    p_rollmean = []

    for n in range(2,7):
        rollmean = TS.rolling(window = n).mean()
        minus_rollmean = TS - rollmean
        minus_rollmean.dropna(inplace = True)
        dftest = adfuller(minus_rollmean)
        p_rollmean.append(dftest[1])
        roll_result[dftest[1]] = rollmean
        roll_param[dftest[1]] = n

    p_rollmean = min(p_rollmean)
    ###
    
    ###
    #Stationarity of curve w/ ewm
    #Checks exponentially weighted mean across iterations of 1-6 periods
    #Selects best DF p value
    decay_result = {}
    decay_param = {}
    p_ewm = []

    for n in range(1,7):
        decay_rollmean = TS.ewm(halflife = n).mean()
        minus_ewm = TS - decay_rollmean
        minus_ewm.dropna(inplace = True)
        dftest = adfuller(minus_ewm)
        p_ewm.append(dftest[1])
        decay_result[dftest[1]] = decay_rollmean
        decay_param[dftest[1]] = n
    
    p_ewm = min(p_ewm)
    ###
    
    ###
    #Stationarity check of differenced dataframe
    #Checks across [1,2,3,4,5,6,7,12,30,365] periods
    diff_result = {}
    diff_param = {}
    p_diff = []
    
    for n in [1,2,3,4,5,6,7,12,30,365]:
        try:
            diff_df = df.diff(periods = n)
            diff_TS = diff_df[diff_df.columns[0]].dropna().values
            dftest = adfuller(diff_TS)
            p_diff.append(dftest[1])
            diff_result[dftest[1]] = diff_TS
            diff_param[dftest[1]] = n
        except:
            continue

    p_diff = min(p_diff)
    
    ###
    #Select lowest P-Value, most staionary curve
    p = min(p_as_is, p_log, p_rollmean, p_ewm, p_diff)
    
    #Return most stationary TS
    if p == p_as_is:
        transform = 'None'
        stat = None
        return TS, p, transform, stat
    
    if p == p_log:
        transform = 'Log'
        stat = None
        return np.log(TS), p, transform, stat
    
    if p == p_rollmean:
        transform = 'Rolling Mean'
        stat = roll_param[p]
        return roll_result[p], p, transform, stat
    
    if p == p_ewm:
        transform = 'EWM'
        stat = decay_param[p]
        return decay_result[p], p, transform, stat
    
    if p == p_diff:
        transform = 'Difference'
        stat = diff_param[p]
        return diff_result[p], p, transform, stat
```

```python
TS, p, transform, stat = best_transform(df_average)
print(f'With a {transform} transform of period {stat}, we scored a {p} on the Dickey-Fuller test of stationarity.')
```

```python
plt.plot(TS)
plt.title('Stationary plot of values, differenced w/ period of 30', fontsize = 20)
```
